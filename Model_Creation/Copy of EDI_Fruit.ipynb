```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
"""
Data Augmentation
--------------------------
rotation_range = rotates the image with the amount of degrees we provide
width_shift_range = shifts the image randomly to the right or left along the width of the image
height_shift range = shifts image randomly to up or below along the height of the image
horizontal_flip = flips the image horizontally
rescale = to scale down the pizel values in our image between 0 and 1
zoom_range = applies random zoom to our object
validation_split = reserves some images to be used for validation purpose
"""

train_datagen = ImageDataGenerator(      rotation_range = 10,
                                         width_shift_range = 0.1,
                                         height_shift_range = 0.1,
                                         horizontal_flip = False,        
                                         rescale = 1./255,
                                         zoom_range = 0.2,
                                         validation_split = 0.2
                                        )#7576765
validation_datagen = ImageDataGenerator(rescale = 1./255,
                                         validation_split = 0.2)
```

```python
"""
Applying data augmentation to the images as we read 
them from their respectivve directories
"""
img_size = 348 #original size of the image
train_generator = train_datagen.flow_from_directory(directory = train_dir,
                                                    target_size = (img_size,img_size),
                                                    batch_size = 16,
                                                    color_mode = "rgb",
                                                    class_mode = "categorical",
                                                    subset = "training"
                                                   )
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
import tensorflow as tf



## Loading VGG16 model
base_model = VGG16(weights= 'imagenet', include_top = False, input_shape=(348,348,3)) #23231231
base_model.trainable = False ## Not trainable weights 12312312312123
```

```python
from tensorflow.keras import layers, models
# With `dilation_rate` as 2.

flatten_layer = layers.Flatten()
dense_layer_1 = layers.Dense(64, activation='relu') #32123213
#dense_layer_2 = layers.Dense(64, activation='relu')
prediction_layer = layers.Dense(22, activation='softmax')


model2 = models.Sequential([
    base_model,
    flatten_layer,
    dense_layer_1,
    #dense_layer_2,
    prediction_layer
])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model2.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'],
)

es = EarlyStopping(monitor='loss', mode='max', patience=8,  restore_best_weights=True)
#y_train = 7 
#model2.fit(train_generator,epochs = 50, validation_data =validation_generator, callbacks=[es])

#es = EarlyStopping(monitor='val_accuracy', mode='max', patience=8,  restore_best_weights=True)
#y_train = 7 
model2.fit(train_generator,epochs = 50)
```
